```python
import numpy as np  # 数据结构
from sklearn.cluster import KMeans  # KMeans聚类
from sklearn.cluster import DBSCAN  # DBSCAN聚类
from sklearn.cluster import AgglomerativeClustering  # 层次聚类
from sklearn.metrics import silhouette_score  # 轮廓系数
```

```python
DataSet = "F:/TCP007/data/model_input/valid_file.npy"  # 数据集
arr=np.load(DataSet,allow_pickle=True)
DataSet=[]
Label=[]
for i in range(len(arr)):
    if arr[i][1]<15:
        continue
    Label.append({'ID':i,'Label':arr[i][1]})
    temp=arr[i][0][0:20].reshape(-1)
    temp=list(temp)
    DataSet.append(temp)
```

```python
len(DataSet)
```

```python
Label
```

```python
models = [("None", None),
          ("Kmeans", KMeans(n_clusters=25)),
          ("DBSCAN", DBSCAN(min_samples=2, eps=1.5)),
          ("AgglomerativeClustering", AgglomerativeClustering(n_clusters=25,linkage="ward"))]
X=np.array(DataSet)
db = models[2][1].fit(X)
labels=db.labels_
for i in range(len(labels)):
    Label[i]['cluster']=labels[i]
```
